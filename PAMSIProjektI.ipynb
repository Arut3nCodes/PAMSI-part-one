
# 1. Zrozumienie biznesowe (Business Understanding)

### Kontekst

Dane pochodzą z programu "Measuring Broadband America" prowadzonego przez amerykańską Federalną Komisję Komunikacji (FCC), który mierzy rzeczywistą wydajność szerokopasmowych usług internetowych.

### Cele analizy:

- Ocena różnicy między deklarowaną a rzeczywistą prędkością internetu

- Identyfikacja czynników wpływających na wydajność łącza

- Porównanie dostawców usług internetowych (ISP)

- Analiza stabilności połączeń

### Metryki:

- Wykrycie istotnych różnic między dostawcami

- Identyfikacja zmiennych najbardziej skorelowanych z wydajnością

- Stworzenie czytelnych wizualizacji pokazujących charakterystykę połączeń
